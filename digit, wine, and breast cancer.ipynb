## MNIST digit classifier 

```python
# (1) import necessary modules

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# load data
mnist = load_digits()

print(dir(mnist))
```

### Understand and explore data

```python
mnist.keys()
```

```python
mnist_data = mnist.data

print(mnist_data.shape)
```

```python
mnist_data[0]
```

```python
mnist_label = mnist.target
print(mnist_label.shape)
mnist_label
```

```python
# print out name of targets
mnist.target_names
```

```python
print(mnist.DESCR)
```

```python
mnist.feature_names
```

```python
import pandas as pd

mnist_df = pd.DataFrame(data=mnist_data, columns = mnist.feature_names)
mnist_df
```

```python
mnist_df["label"] = mnist.target
mnist_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mnist_data,
                                                    mnist_label,
                                                    test_size=0.2,
                                                    random_state=8)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
y_train, y_test
```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
print(decision_tree._estimator_type)
```

```python
decision_tree.fit(X_train, y_train)
```

```python
y_pred = decision_tree.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
print(classification_report(y_test, y_pred))
```

```python
# RandomForest Model
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# SVM model

from sklearn import svm
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
y_pred

print(classification_report(y_test, y_pred))
```

```python
#SGD model

from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
y_pred

print(classification_report(y_test, y_pred))
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter = 10000)

print(logistic_model._estimator_type)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
y_pred

print(classification_report(y_test, y_pred))
```
